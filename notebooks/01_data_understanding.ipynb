```python
import pandas as pd
import numpy as np

df = pd.read_csv('../data/Online Retail.csv', sep=';')

print(f"Всего строк: {df.shape[0]}, Всего столбцов: {df.shape[1]}")

print("\nИзначальные типы:")

for col in df.columns:
    dtype = df[col].dtype
    description = ""
    match col:
        case 'InvoiceNo':
            description = "Номер заказа"
        case 'StockCode':
            description = "Код товара"
        case 'Description':
            description = "Название/описание товара"
        case 'Quantity':
            description = "Количество товара в заказе (может быть отрицательным для возвратов)"
        case 'InvoiceDate':
            description = "Дата и время создания заказа"
        case 'UnitPrice':
            description = "Цена за единицу товара (в фунтах стерлингов)"
        case 'CustomerID':
            description = "Уникальный идентификатор клиента"
        case 'Country':
            description = "Страна доставки/клиента"
    print(f"{col:<15} {str(dtype):<10} - {description}")

print("\nПреобразование типов к более подходящим:")
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'], format='%d.%m.%Y %H:%M', errors='coerce')
print("InvoiceDate: преобразован в datetime")
df['UnitPrice'] = df['UnitPrice'].str.replace(',', '.').astype(float)
print("UnitPrice: преобразован в float")
df['CustomerID'] = df['CustomerID'].astype('Int64')
print("CustomerID: преобразован в Int64")
df['InvoiceNo'] = df['InvoiceNo'].astype('category')
print("InvoiceNo: преобразован в category")
df['StockCode'] = df['StockCode'].astype('category')
print("StockCode: преобразован в category")
df['Country'] = df['Country'].astype('category')
print("Country: преобразован в category")

print("\nИтоговые типы данных:")
print(df.dtypes)
```

# 3. Аудит качества данных

```python
import matplotlib.pyplot as plt
import seaborn as sns

print("Пропуски (Missing Values)\n")

missing_data = pd.DataFrame({
    'Количество пропусков': df.isnull().sum(),
    'Доля пропусков, %': (df.isnull().sum() / len(df) * 100).round(2)
})
print(missing_data)

print("\nНаиболее проблемные признаки:")
print(f"- CustomerID: {missing_data.loc['CustomerID', 'Доля пропусков, %']}% пропусков")
print(f"- Description: {missing_data.loc['Description', 'Доля пропусков, %']}% пропусков")

print("\nВывод:")
print("CustomerID: ~25% пропусков - критично для анализа клиентского поведения")
print("Description: 0.27% пропусков - не критично")
print("Остальные признаки: пропусков нет")

print("\nДубликаты")

full_duplicates = df.duplicated(keep = "first").sum()
print(f"Дубликатов строк: {full_duplicates}")

print("Типические проблемы значений\n")

print("Числовые признаки:")
numeric_cols = ['Quantity', 'UnitPrice']
for col in numeric_cols:
    print(f"\n{col}:")
    print(f"  Min: {df[col].min()}, Max: {df[col].max()}")
    print(f"  Mean: {df[col].mean():.2f}, Std: {df[col].std():.2f}")
    print(f"  Медиана: {df[col].median():.2f}")
    
    if col == 'Quantity':
        negative = (df[col] < 0).sum()
        print(f"  Отрицательные значения (возвраты): {negative}")
    
    if col == 'UnitPrice':
        zero_price = (df[col] == 0).sum()
        negative_price = (df[col] < 0).sum()
        print(f"  Нулевые цены: {zero_price}")
        print(f"  Отрицательные цены: {negative_price}")

print("\nКатегориальные признаки:")
cat_cols = ['InvoiceNo', 'StockCode', 'Description', 'Country']
for col in cat_cols:
    unique_count = df[col].nunique()
    print(f"\n{col}:")
    print(f"  Уникальных значений: {unique_count}")
    
    if col == 'Description':        
        leading_spaces = df[col].astype(str).str.startswith(' ').sum()
        trailing_spaces = df[col].astype(str).str.endswith(' ').sum()
        print(f"  Значений с начальными пробелами: {leading_spaces}")
        print(f"  Значений с конечными пробелами: {trailing_spaces}")

print("Выбросы (Outliers)\n")

outliers_cols = ['Quantity','UnitPrice']
for col in outliers_cols:
    print(f"Анализ выбросов для {col}:")
    
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    outliers_iqr = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
    print(f"Метод IQR (границы: [{lower_bound:.1f}, {upper_bound:.1f}]):")
    print(f"  Выбросов: {len(outliers_iqr)} ({len(outliers_iqr)/len(df)*100:.2f}%)")

    fig, axes = plt.subplots(1, 2, figsize=(12, 5))
    
    axes[0].boxplot(df[col].dropna())
    axes[0].set_title(f'Boxplot {col}')
    axes[0].set_ylabel(col)
    
    axes[1].scatter(range(len(df)), df[col], alpha=0.5, s=1)
    axes[1].axhline(y=upper_bound, color='r', linestyle='--', label=f'Верхняя граница ({upper_bound:.1f})')
    axes[1].axhline(y=lower_bound, color='r', linestyle='--', label=f'Нижняя граница ({lower_bound:.1f})')
    axes[1].set_title(f'Scatter Plot {col}')
    axes[1].set_xlabel('Индекс записи')
    axes[1].set_ylabel(col)
    axes[1].legend()
    
    if col == 'Quantity':
        axes[1].set_ylim([-1000, 1000])
    elif col == 'UnitPrice':
        axes[1].set_ylim([-50, 500])
    
    plt.tight_layout()
    plt.show()
```

## Пропуски и их природа (MCAR / MAR / MNAR)

#### Рассмотрим признак CustomerID  
 Гипотеза 1: Пропуски относятся к MCAR  
Вероятность пропуска не связана ни с величиной пропущенного значения признака, ни с какими-либо другими признаками транзакции.  

 Гипотеза 2: Пропуски относятся к MAR  
Вероятность пропуска связана с другими признаками транзации, но не зависит от величины пропущенного значения признака.  

 Гипотеза 3: Пропуски относятся к MNAR  
Вероятность пропуска зависит от величины пропущенного значения.  


```python
import matplotlib.pyplot as plt

print('Всего строк:', len(df))
print('Пропуски в CustomerID:', df['CustomerID'].isnull().sum())

mask_missing = df['CustomerID'].isnull()
mask_observed = ~mask_missing

print('\nСтрок с пропусками:', mask_missing.sum())
print('Строк без пропусков:', mask_observed.sum())

print('\n=== Гипотеза MCAR ===')
print('Подтверждение: нет различий в распределениях UnitPrice, Quantity, Country')

print('UnitPrice среднее без пропусков:', df.loc[mask_observed, 'UnitPrice'].mean().round(2))
print('UnitPrice среднее с пропусками:', df.loc[mask_missing, 'UnitPrice'].mean().round(2))

print('Quantity среднее без пропусков:', df.loc[mask_observed, 'Quantity'].mean().round(2))
print('Quantity среднее с пропусками:', df.loc[mask_missing, 'Quantity'].mean().round(2))
print('Гипотеза MCAR отвергается, средние значения с пропусками и без них значительно отличаются')

print('\n=== Гипотеза MAR ===')
print('Подтверждение: зависимость от UnitPrice/Country')

pct_missing_country = df.groupby('Country')['CustomerID'].apply(lambda x: x.isnull().mean() * 100).sort_values(ascending=False).head(10)
print('Топ-10 стран по % пропусков:')
print(pct_missing_country.round(2))
print('Гипотеза MAR принимается, процент пропусков зависит от признака Country')

print('\n=== Гипотеза MNAR ===')
print('Подтверждение: пропуски связаны с поведением клиента (разовые покупки, возвраты)')
df['is_return'] = df['Quantity'] < 0
df['is_small_ticket'] = (df['UnitPrice'] * abs(df['Quantity'])) < 1

print('\n1. Анализ возвратов (Quantity < 0):')
return_missing = df.groupby('is_return')['CustomerID'].apply(lambda x: x.isna().mean() * 100).round(2)
print(f'Доля пропусков в обычных покупках: {return_missing[False]}%')
print(f'Доля пропусков в возвратах: {return_missing[True]}%')

print('Вывод: есть связь')

print('\n2. Анализ очень маленьких чеков (< 1 фунт):')
small_missing = df.groupby('is_small_ticket')['CustomerID'].apply(lambda x: x.isna().mean() * 100).round(2)
print(f'Доля пропусков в обычных чеках: {small_missing[False]}%')
print(f'Доля пропусков в маленьких чеках: {small_missing[True]}%')

print('В возвратах значительно меньше пропусков, в очень маленьких покупках в два раза больше пропусков. Оба факта могут быть объяснены поведением покупателя, MNAR возможна.')
```

### Когда факт пропуска несет информацию?
Факт пропуска несёт информацию при MAR/MNAR, когда отсутствие значения коррелирует с другими признаками.

### Почему значимость индикатора в модели - это сигнал о неправильной или грубой импутации?
Значимость индикатора пропуска указывает на то, что сам факт отсутствия данных несет предсказательную силу, которая не была учтена при заполнении.

### Пример признака, для которого индикатор пропуска полезен, бесполезен, опасен.
Полезен: при MNAR, когда пропуск несет информацию
Бесполезен: при MCAR, когда пропуск не несет информации
Опасен: когда пропуск - неточный индикатор информации, на основании которого модель может делать неверные выводы

## Статистики
Логарифмирование предпочтительно, когда данные имеют экспоненциальный рост. Работает только с положительными числами.

Корень предпочтителен при умеренной асимметрии и наличии нулей. Сохраняет единицы измерения.

Винзоризация предпочтительна, когда важны исходные значения, но нужно убрать влияние выбросов. Не меняет форму распределения радикально.
Влияние подходов:

|  | Логарифмирование | Корень | Винзоризация |
|--------|------------------|--------|--------------|
| Интерпретируемость | Сложная | Средняя | Высокая |
| Линейные модели | Сильно улучшает, если связь мультипликативна | Умеренно улучшает | Улучшает за счет снижения влияния выбросов |
| Визуализация | Лучше видно крайние значения | Умеренное сжатие правого края | Почти не меняет форму, обрезает края |

```python
df_clean = df[df['UnitPrice'] > 0].copy()
print('Арифметическое среднее:', df_clean['UnitPrice'].mean().round(2))
print('Медиана:', df_clean['UnitPrice'].median().round(2))
print('Геометрическое среднее:', np.exp(np.log(df_clean['UnitPrice']).mean()).round(2))
```

Средние значения различаются из-за способа вычисления. Арифметическое среднее чувствительно к верхним выбросам, медиана и геометрическое среднее же лучше отображают среднюю покупку.
Лучше всего отражает среднюю покупку медиана, т.к. не учитывает скос из-за небольшого количества дорогих заказов.

## Визуализация как инструмент мышления

Рассмотрим визуализацию количества заказов по странам в формате гистограммы. В качестве худшего выбора визуализации можно привести круговую диаграмму. При её использовании 90% круга заняла бы Великобритания, визуальное сравнение других стран было бы затруднено. Пользователь мог бы сделать вывод, что все рынки кроме Великобритании одинаково невостребованы и не приносят заказов.

Рассмотрим пару признаков Quantity и UnitPrice. Для них можно построить визуализации в виде точечной диаграммы и ящика с усами. 
Точечная диаграмма подчеркивает индивидуальные пары значений и плотность точек. Гипотеза: "Дорогие товары берут по 1 шт"
Ящик с усами подчеркивает центр и разброс цен в каждой группе количества, позволяет сравнивать медианы в каждой группе. Гипотеза: "При крупных заказах средняя цена падает в n раз"
